## Example spectrograms

```python
import numpy as np
import matplotlib.pyplot as plt
import librosa
import librosa.display
```

```python
wav, sr = librosa.load(f'data/train_audio/{meta_df.iloc[37]["filename"]}')
```

```python

# Generate example waveform
sr = 22050  # Sampling rate
# t = np.linspace(0, 1, sr)
# Complex waveform: combination of sine waves with different frequencies
# waveform = np.sin(2 * np.pi * 3 * t) + 0.5 * np.sin(2 * np.pi * 7 * t) + 0.25 * np.sin(2 * np.pi * 13 * t)
waveform = wav[sr * 0:sr * 5]

# Compute the spectrogram
S = librosa.stft(waveform)
S_db = librosa.amplitude_to_db(np.abs(S), ref=np.max)

# Compute the mel spectrogram
S_mel = librosa.feature.melspectrogram(y=waveform, sr=sr, n_mels=128, fmax=8000)
S_mel_db = librosa.power_to_db(S_mel, ref=np.max)

# Plotting
plt.figure(figsize=(15, 3))

# Plot waveform
plt.subplot(1, 3, 1)
librosa.display.waveshow(waveform, sr=sr)
plt.title('Waveform')

# Plot spectrogram
plt.subplot(1, 3, 2)
librosa.display.specshow(S_db, sr=sr, x_axis='time', y_axis='log')
plt.colorbar(format='%+2.0f dB')
plt.title('Spectrogram')

# Plot mel spectrogram
plt.subplot(1, 3, 3)
librosa.display.specshow(S_mel_db, sr=sr, x_axis='time', y_axis='mel', fmax=8000)
plt.colorbar(format='%+2.0f dB')
plt.title('Mel Spectrogram')

plt.tight_layout()
plt.show()
```
